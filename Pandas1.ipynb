```python
import pandas as pd
df1 = pd.DataFrame(
    [
        (1, 345, 'B', True),
                    (2, 100, 'C', False),
                    (3, 300, 'B', False),
                    (4, 151, 'A', True),
                    (5, 170, 'B', True),
                    (6, 187, 'A', False),
                   ],
                   columns = ['id', 'value', 'colC', 'coID']
                  )
df2 = pd.DataFrame(
    [
        (1,111,10.2,3),
        (9,56,10,1),
        (3,567,19.1,4),
        (11,98,2.1,1),
    ],
    columns = ['id','value','colE','colF']
)
```

```python
df1
```

```python
df2
```

```python
#merge data 

df1.merge(df2, on = 'id')
```

```python
#left - join
#it is taking left join(df1) values and comman element from df2
df1.merge(df2, on = 'id', how = 'left')
```

```python
#right join 
#it is taking right join(df2) values and comman element from df2
df1.merge(df2, on = 'id', how = 'right')
```

```python
#outer join
#all the element merge together 
df1.merge(df2, on = 'id', how = 'outer')
```
